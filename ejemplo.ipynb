# P3 (10 puntos)

Considere el archivo censo.csv visto en clases. Obtenga la cantidad de hombres y mujeres por separado de la comuna de Arica, Santiago, Concepción y Talca. (5 puntos)

Y luego realice un gráfico de barras con la cantidad de hombres y mujeres por las comunas pedidas. Recuerde distinguir entre hombres y mujeres (por color).(5 puntos)

Usar sólo lenguaje Python con **PANDAS** y no llamadas al systema con sys o mezclar bash

Si desea leer directo el archivo, le dejo link directo: https://raw.githubusercontent.com/malegria01/AD/main/dataset/censo.csv

````python
##PANDAS----
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/malegria01/AD/main/dataset/censo.csv") # se crea el data frame

#df = pd.DataFrame(data)

Total_comunal = df[df['Edad'] == 'Total Comunal']

arica = Total_comunal[Total_comunal['NOMBRE COMUNA'] == 'ARICA'][['NOMBRE COMUNA', 'Mujeres', 'Hombres']]
talca = Total_comunal[Total_comunal['NOMBRE COMUNA'] == 'TALCA'][['NOMBRE COMUNA', 'Mujeres', 'Hombres']]
santiago = Total_comunal[Total_comunal['NOMBRE COMUNA'] == 'SANTIAGO'][['NOMBRE COMUNA', 'Mujeres', 'Hombres']]
concepcion = Total_comunal[Total_comunal['NOMBRE COMUNA'] == 'CONCEPCIÓN'][['NOMBRE COMUNA', 'Mujeres', 'Hombres']]


dataFrameCombinado = pd.concat([arica,talca,santiago,concepcion])
final = dataFrameCombinado.set_index('NOMBRE COMUNA')
final
````

````python

import matplotlib.pyplot as plt

s_plot = final.plot.barh(fontsize=12,figsize=(15,10))
s_plot.set_ylabel("Región", fontsize=15)
s_plot.set_xlabel("Cantidad de Mujeres en millones",fontsize=15)
s_plot.set_title("Cantidad de mujeres totales por Región", fontsize=20)
s_plot.invert_yaxis()

#tips: si queremos guardar la figura siempre es antes de plt.show
plt.tight_layout()
#plt.savefig('plot1.png', format='png', dpi=600)

plt.show()
````

Cree un dataframe con el total de  la población y con el total de casos confirmados por región
